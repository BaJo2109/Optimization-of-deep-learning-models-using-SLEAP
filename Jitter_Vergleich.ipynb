```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter


WINDOW = 7
POLYORDER = 2



# Nase Koordinaten extrahieren
big_x = Sigma_groß["nose.x"].to_numpy()
big_y = Sigma_groß["nose.y"].to_numpy()

# Frame-zu-Frame Differenzen
big_dx = np.diff(big_x)
big_dy = np.diff(big_y)

# Δr raw
dr_big_raw = np.sqrt(big_dx**2 + big_dy**2)

# Interpolation (falls NaNs existieren)
big_x_interp = pd.Series(big_x).interpolate(limit_direction="both").to_numpy()
big_y_interp = pd.Series(big_y).interpolate(limit_direction="both").to_numpy()

# Savitzky-Golay Glättung
big_x_smooth = savgol_filter(big_x_interp, window_length=WINDOW, polyorder=POLYORDER)
big_y_smooth = savgol_filter(big_y_interp, window_length=WINDOW, polyorder=POLYORDER)

# Differenzen nach Glättung
big_dx_s = np.diff(big_x_smooth)
big_dy_s = np.diff(big_y_smooth)

# Δr smooth
dr_big_s = np.sqrt(big_dx_s**2 + big_dy_s**2)

# Statistik
big_median_raw = float(np.nanmedian(dr_big_raw))
big_p95_raw = float(np.nanpercentile(dr_big_raw, 95))

big_median_s = float(np.nanmedian(dr_big_s))
big_p95_s = float(np.nanpercentile(dr_big_s, 95))

big_median_reduction = 100 * (big_median_raw - big_median_s) / big_median_raw
big_p95_reduction = 100 * (big_p95_raw - big_p95_s) / big_p95_raw


small_x = Sigma_klein["nose.x"].to_numpy()
small_y = Sigma_klein["nose.y"].to_numpy()

small_dx = np.diff(small_x)
small_dy = np.diff(small_y)

dr_small_raw = np.sqrt(small_dx**2 + small_dy**2)

small_x_interp = pd.Series(small_x).interpolate(limit_direction="both").to_numpy()
small_y_interp = pd.Series(small_y).interpolate(limit_direction="both").to_numpy()

small_x_smooth = savgol_filter(small_x_interp, window_length=WINDOW, polyorder=POLYORDER)
small_y_smooth = savgol_filter(small_y_interp, window_length=WINDOW, polyorder=POLYORDER)

small_dx_s = np.diff(small_x_smooth)
small_dy_s = np.diff(small_y_smooth)

dr_small_s = np.sqrt(small_dx_s**2 + small_dy_s**2)

small_median_raw = float(np.nanmedian(dr_small_raw))
small_p95_raw = float(np.nanpercentile(dr_small_raw, 95))

small_median_s = float(np.nanmedian(dr_small_s))
small_p95_s = float(np.nanpercentile(dr_small_s, 95))

small_median_reduction = 100 * (small_median_raw - small_median_s) / small_median_raw
small_p95_reduction = 100 * (small_p95_raw - small_p95_s) / small_p95_raw



print("\nBig Sigma:")
print("Median raw:", big_median_raw)
print("Median smooth:", big_median_s)
print("Median reduction %:", big_median_reduction)
print("p95 raw:", big_p95_raw)
print("p95 smooth:", big_p95_s)
print("p95 reduction %:", big_p95_reduction)

print("\nSmall Sigma:")
print("Median raw:", small_median_raw)
print("Median smooth:", small_median_s)
print("Median reduction %:", small_median_reduction)
print("p95 raw:", small_p95_raw)
print("p95 smooth:", small_p95_s)
print("p95 reduction %:", small_p95_reduction)


plt.figure(figsize=(12, 5))

plt.plot(dr_big_raw, label="big Sigma raw", alpha=0.4)
plt.plot(dr_big_s, label="big Sigma smooth", linewidth=2)

plt.plot(dr_small_raw, label="small Sigma raw", alpha=0.4)
plt.plot(dr_small_s, label="small Sigma smooth", linewidth=2)

plt.xlabel("Frame")
plt.ylabel("Δr (Pixel / Frame)")
plt.title("Jitter-Vergleich (Δr Zeitreihe)")
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 5))

bins = 100

plt.hist(dr_big_raw, bins=bins, alpha=0.4, label="big Sigma raw", density=True)
plt.hist(dr_big_s, bins=bins, alpha=0.6, label="big Sigma smooth", density=True)

plt.hist(dr_small_raw, bins=bins, alpha=0.4, label="small Sigma raw", density=True)
plt.hist(dr_small_s, bins=bins, alpha=0.6, label="small Sigma smooth", density=True)

plt.xlabel("Δr (Pixel / Frame)")
plt.ylabel("Dichte")
plt.title("Jitter-Verteilung (Histogramm)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def stats(dr_raw, dr_s, thr=None):
    if thr is not None:
        m = dr_raw < thr
        dr_raw = dr_raw[m]
        dr_s = dr_s[m]

    med_r = np.nanmedian(dr_raw)
    p95_r = np.nanpercentile(dr_raw, 95)
    med_s = np.nanmedian(dr_s)
    p95_s = np.nanpercentile(dr_s, 95)

    return {
        "med_raw": med_r,
        "p95_raw": p95_r,
        "med_s": med_s,
        "p95_s": p95_s,
        "med_red_%": 100 * (med_r - med_s) / med_r,
        "p95_red_%": 100 * (p95_r - p95_s) / p95_r,
    }

# Optional: nur "kleine Bewegungen" als Jitter betrachten
THR = None     # z.B. 10 setzen, wenn du echte Bewegungen rausfiltern willst

big = stats(dr_big_raw, dr_big_s, thr=THR)
small = stats(dr_small_raw, dr_small_s, thr=THR)

fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True, sharey=True)

# --- Big sigma ---
axes[0].plot(dr_big_raw, label="raw", alpha=0.4)
axes[0].plot(dr_big_s, label="smooth", linewidth=2)
axes[0].set_ylabel("Δr (px/frame)")
axes[0].legend()
axes[0].set_title(
    f"big σ | median {big['med_raw']:.2f}->{big['med_s']:.2f} ({big['med_red_%']:.1f}%)"
    f" | p95 {big['p95_raw']:.2f}->{big['p95_s']:.2f} ({big['p95_red_%']:.1f}%)"
)

# --- Small sigma ---
axes[1].plot(dr_small_raw, label="raw", alpha=0.4)
axes[1].plot(dr_small_s, label="smooth", linewidth=2)
axes[1].set_xlabel("Frame")
axes[1].set_ylabel("Δr (px/frame)")
axes[1].legend()
axes[1].set_title(
    f"small σ | median {small['med_raw']:.2f}->{small['med_s']:.2f} ({small['med_red_%']:.1f}%)"
    f" | p95 {small['p95_raw']:.2f}->{small['p95_s']:.2f} ({small['p95_red_%']:.1f}%)"
)

plt.tight_layout()
plt.show()

print("Jitter summary:")
print("big σ  :", big)
print("small σ:", small)
```
