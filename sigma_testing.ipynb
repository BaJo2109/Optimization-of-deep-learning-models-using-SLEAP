```python
import pandas as pd
import numpy as np
#Load the CSV Sleap files
self_labeled = pd.read_csv(r'C:\Users\jrall\Desktop\Bachelorarbeit\Ergebnisse\Sigma\CSV\self_labeled.csv')
big_sigma = pd.read_csv(r'C:\Users\jrall\Desktop\Bachelorarbeit\Ergebnisse\Sigma\CSV\big_sigma.csv')
small_sigma = pd.read_csv(r'C:\Users\jrall\Desktop\Bachelorarbeit\Ergebnisse\Sigma\CSV\small_sigma.csv')
```

### Calculate the euklidian distance to my own labeling and add it to a new file

```python
self_labeled
```

```python
#add the euclidean distance to self labeled between x and y nose, x and y neck, x and y tail base
self_labeled['ED_big_sigma_nose'] = np.sqrt((self_labeled['nose.x'] - big_sigma['nose.x'])**2 + (self_labeled['nose.y'] - big_sigma['nose.y'])**2)
self_labeled['ED_big_sigma_neck'] = np.sqrt((self_labeled['neck.x'] - big_sigma['neck.x'])**2 + (self_labeled['neck.y'] - big_sigma['neck.y'])**2)
self_labeled['ED_big_sigma_tail_base'] = np.sqrt((self_labeled['tail_base.x'] - big_sigma['tail_base.x'])**2 + (self_labeled['tail_base.y'] - big_sigma['tail_base.y'])**2)
self_labeled['ED_small_sigma_nose'] = np.sqrt((self_labeled['nose.x'] - small_sigma['nose.x'])**2 + (self_labeled['nose.y'] - small_sigma['nose.y'])**2)
self_labeled['ED_small_sigma_neck'] = np.sqrt((self_labeled['neck.x'] - small_sigma['neck.x'])**2 + (self_labeled['neck.y'] - small_sigma['neck.y'])**2)
self_labeled['ED_small_sigma_tail_base'] = np.sqrt((self_labeled['tail_base.x'] - small_sigma['tail_base.x'])**2 + (self_labeled['tail_base.y'] - small_sigma['tail_base.y'])**2)
self_labeled
```

```python
#MEAN values of nose, neck and tail base distances of big sigma and small sigma and count the nan values of the distances
mean_ED_big_sigma_nose = self_labeled['ED_big_sigma_nose'].mean()
mean_ED_big_sigma_neck = self_labeled['ED_big_sigma_neck'].mean()
mean_ED_big_sigma_tail_base = self_labeled['ED_big_sigma_tail_base'].mean()
mean_ED_small_sigma_nose = self_labeled['ED_small_sigma_nose'].mean()
mean_ED_small_sigma_neck = self_labeled['ED_small_sigma_neck'].mean()
mean_ED_small_sigma_tail_base = self_labeled['ED_small_sigma_tail_base'].mean()
print("Mean Euclidean Distance between Self Labeled and Big Sigma:")
print("Nose in px: ", mean_ED_big_sigma_nose)
print("Neck in px: ", mean_ED_big_sigma_neck)
print("Tail Base in px: ", mean_ED_big_sigma_tail_base)
print("Mean Euclidean Distance between Self Labeled and Small Sigma:")
print("Nose in px: ", mean_ED_small_sigma_nose)
print("Neck in px: ", mean_ED_small_sigma_neck)
print("Tail Base in px: ", mean_ED_small_sigma_tail_base)
print("Count of NaN values in Euclidean Distance between Self Labeled and Big Sigma:")
print("Nose in px: ", self_labeled['ED_big_sigma_nose'].isna().sum())
print("Neck in px: ", self_labeled['ED_big_sigma_neck'].isna().sum())
print("Tail Base in px: ", self_labeled['ED_big_sigma_tail_base'].isna().sum())
print("Count of NaN values in Euclidean Distance between Self Labeled and Small Sigma:")
print("Nose in px: ", self_labeled['ED_small_sigma_nose'].isna().sum())
print("Neck in px: ", self_labeled['ED_small_sigma_neck'].isna().sum())
print("Tail Base in px: ", self_labeled['ED_small_sigma_tail_base'].isna().sum())

#Calculate the in centimeters assuming 1 pixel = 0.045 cm and print in cm and mm
print("Mean Euclidean Distance between Self Labeled and Big Sigma in cm:")
print("Nose in cm: ", mean_ED_big_sigma_nose * 0.045, "cm", mean_ED_big_sigma_nose * 0.045 * 10, "mm")
print("Neck in cm: ", mean_ED_big_sigma_neck * 0.045, "cm", mean_ED_big_sigma_neck * 0.045 * 10, "mm")
print("Tail Base in cm: ", mean_ED_big_sigma_tail_base * 0.045, "cm", mean_ED_big_sigma_tail_base * 0.045 * 10, "mm")
print("Mean Euclidean Distance between Self Labeled and Small Sigma in cm:")
print("Nose in cm: ", mean_ED_small_sigma_nose * 0.045, "cm", mean_ED_small_sigma_nose * 0.045 * 10, "mm")
print("Neck in cm: ", mean_ED_small_sigma_neck * 0.045, "cm", mean_ED_small_sigma_neck * 0.045 * 10, "mm")
print("Tail Base in cm: ", mean_ED_small_sigma_tail_base * 0.045, "cm", mean_ED_small_sigma_tail_base * 0.045 * 10, "mm")

#Calculate standart error of the mean for the distances
sem_ED_big_sigma_nose = self_labeled['ED_big_sigma_nose'].sem()
sem_ED_big_sigma_neck = self_labeled['ED_big_sigma_neck'].sem()
sem_ED_big_sigma_tail_base = self_labeled['ED_big_sigma_tail_base'].sem()
sem_ED_small_sigma_nose = self_labeled['ED_small_sigma_nose'].sem()
sem_ED_small_sigma_neck = self_labeled['ED_small_sigma_neck'].sem()
sem_ED_small_sigma_tail_base = self_labeled['ED_small_sigma_tail_base'].sem()
print("Standard Error of the Mean for Euclidean Distance between Self Labeled and Big Sigma:")
print("Nose in px: ", sem_ED_big_sigma_nose)
print("Neck in px: ", sem_ED_big_sigma_neck)
print("Tail Base in px: ", sem_ED_big_sigma_tail_base)
print("Standard Error of the Mean for Euclidean Distance between Self Labeled and Small Sigma:")
print("Nose in px: ", sem_ED_small_sigma_nose)
print("Neck in px: ", sem_ED_small_sigma_neck)
print("Tail Base in px: ", sem_ED_small_sigma_tail_base)

```

```python
#Plot two histograms showing the distribution of big and small sigma distances  for nose in two subplots, the ticks for x axis is the same in both histograms

import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.hist(self_labeled['ED_big_sigma_nose'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax1.set_title('Distribution of Euclidean Distance - Big Sigma (Nose)')
ax1.set_xlabel('Euclidean Distance (px)')
ax1.set_ylabel('Frequency')
ax1.set_xlim(0, 20)
ax2.hist(self_labeled['ED_small_sigma_nose'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax2.set_title('Distribution of Euclidean Distance - Small Sigma (Nose)')
ax2.set_xlabel('Euclidean Distance (px)')
ax2.set_ylabel('Frequency')
ax2.set_xlim(0, 20)
plt.tight_layout()
plt.show()

#Add the plots to compare neck big and small sigma
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.hist(self_labeled['ED_big_sigma_neck'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax1.set_title('Distribution of Euclidean Distance - Big Sigma (Neck)')
ax1.set_xlabel('Euclidean Distance (px)')
ax1.set_ylabel('Frequency')
ax1.set_xlim(0, 20)
ax2.hist(self_labeled['ED_small_sigma_neck'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax2.set_title('Distribution of Euclidean Distance - Small Sigma (Neck)')
ax2.set_xlabel('Euclidean Distance (px)')
ax2.set_ylabel('Frequency')
ax2.set_xlim(0, 20)
plt.tight_layout()
plt.show()

#Add the plots to compare tail base big and small sigma
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.hist(self_labeled['ED_big_sigma_tail_base'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax1.set_title('Distribution of Euclidean Distance - Big Sigma (Tail Base)')
ax1.set_xlabel('Euclidean Distance (px)')
ax1.set_ylabel('Frequency')
ax1.set_xlim(0, 20)
ax2.hist(self_labeled['ED_small_sigma_tail_base'].dropna(), bins=10, edgecolor='black', color='blue', alpha=1)
ax2.set_title('Distribution of Euclidean Distance - Small Sigma (Tail Base)')
ax2.set_xlabel('Euclidean Distance (px)')
ax2.set_ylabel('Frequency')
ax2.set_xlim(0, 20)
plt.tight_layout()
plt.show()
```
